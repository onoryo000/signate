```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
train = pd.read_csv("train.csv", index_col=0) # 学習用データ
test = pd.read_csv("test.csv", index_col=0)   # 評価用データ
sample_submit = pd.read_csv("sample_submit.csv", index_col=0, header=None) # 応募用サンプルファイル
```

```python
pd.get_dummies(train).corrwith(train["charges"])
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
train["age_bmi"] = train["age"] * train["bmi"]
test["age_bmi"] = test["age"] * test["bmi"]


train["age_male"] = train["age"] / (train["sex_male"]+1)
test["age_male"] = test["age"] / (test["sex_male"]+1)

train["age_region_northwest"] = train["age"] / (train["region_northwest"]+1)
test["age_region_northwest"] = test["age"] / (test["region_northwest"]+1)

train["age_region_southeast "] = train["age"] / (train["region_southeast"]+1)
test["age_region_southeast "] = test["age"] / (test["region_southeast"]+1)

train["age_squared"] = train["age"]**2
test["age_squared"] = test["age"]**2
train["bmi_log"] = np.log(train["bmi"] + 1)
test["bmi_log"] = np.log(test["bmi"] + 1)


# train["smoker_yes_region_southeast"]=train["smoker_yes"]*train["region_southeast"]
# test["smoker_yes_region_southeast"] = test["smoker_yes"] *test["region_southeast"]

# train["smoke_bmi_age"] = train["smoker_yes"] * train["bmi"]*train["age"] 
# test["smoke_bmi_age"] = test["smoker_yes"] * test["bmi"]*test["age"]
```

```python
from sklearn.preprocessing import StandardScaler

# 標準化の対象となる特徴量を選択
selected_features = ["age", "bmi","age_bmi","age_squared","bmi_log","age_male","region_northwest","region_southeast"]

# StandardScalerのインスタンスを作成
scaler = StandardScaler()
# 選択した特徴量の平均値と標準偏差を計算し、標準化を実行
train[selected_features] = scaler.fit_transform(train[selected_features])
test[selected_features] = scaler.transform(test[selected_features])
```

```python
import lightgbm as lgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

```

```python
# データセットをトレーニングセットとテストセットに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
# LightGBM用のデータセットを作成
train_data = lgb.Dataset(X_train, label=y_train)



# from sklearn.model_selection import GridSearchCV
# import lightgbm as lgb

# # パラメータグリッドの定義
# param_grid = {
#     'num_leaves': [10, 20, 30],
#     'learning_rate': [0.1, 0.01, 0.001],
#     'max_bin': [ 200, 250, 300],
#     'max_depth': [0,1,2],
#     'min_data_in_leaf' : [10,20,30]
# }

# # LightGBM分類器のインスタンス化
# lgb_model = lgb.LGBMClassifier(objective='multiclass', num_class=3)

# # グリッドサーチによるパラメータ探索
# grid_search = GridSearchCV(estimator=lgb_model,
#                            param_grid=param_grid,
#                            scoring='f1_macro',
#                            cv=5)

# # モデルの学習と最適なパラメータの探索
# grid_search.fit(X_train, y_train)

# # 最適なパラメータの取得
# best_params = grid_search.best_params_


# パラメータの設定
params = {'objective': 'multiclass',
    'num_class': 3,
    'lambda_l1': 0.1,
    'lambda_l2': 0.1,
    'learning_rate': 0.1,
    'max_bin': 300,
    'max_depth': 2,
    'min_data_in_leaf': 30,
    'num_leaves': 10}

# params.update(best_params)

params
```

```python
# best_f1 = 0.0
# best_num_boost_round = 0
# sumf1=0.0
# for num_boost_round in range(1,101):
#     # モデルの学習
#     model = lgb.train(params, train_data, num_boost_round)
#     # テストデータの予測
#     y_pred = model.predict(X_test)
#     y_pred_labels = y_pred.argmax(axis=1)
#     # 精度の評価
#     from sklearn.metrics import f1_score
#     f1_macro = f1_score(y_test, y_pred_labels, average='macro')
#     #print("F1 Macro Score:", f1_macro)
#         # 最も高いF1スコアとその時のnum_boost_roundを更新
#     sumf1+=f1_macro
#     if f1_macro > best_f1:
#         best_f1 = f1_macro
#         best_num_boost_round = num_boost_round
    
# print(sumf1/100)
# # 最適なnum_boost_roundを表示
# print("Best num_boost_round:", best_num_boost_round)
# print("Best F1 score:", best_f1)


model = lgb.train(params, train_data, num_boost_round=88)
# テストデータの予測
y_pred = model.predict(X_test)
y_pred_labels = y_pred.argmax(axis=1)
# 精度の評価
from sklearn.metrics import f1_score
f1_macro = f1_score(y_test, y_pred_labels, average='macro')
print("F1 Macro Score:", f1_macro)

# 評価データの予測
pred = model.predict(test)
# 予測結果をクラスラベルに変換
pred_labels = pred.argmax(axis=1)
```

```python
sample_submit[1] = pred_labels
sample_submit.to_csv('submit.csv', header=None, sep=',')
```
